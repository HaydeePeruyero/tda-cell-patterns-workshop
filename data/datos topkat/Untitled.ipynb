```python
import pandas as pd

# Leer archivo
df = pd.read_csv("datos_50_FA_50_NOFA.csv")

# Guardar PID original por seguridad
df["PID_original"] = df["PID"]

# Crear nuevos PID consecutivos
pid_map = {
    pid: i + 1
    for i, pid in enumerate(sorted(df["PID"].unique()))
}

df["PID"] = df["PID"].map(pid_map)

# Si también quieres actualizar la columna id
if "id" in df.columns:
    df["id_original"] = df["id"]
    df["id"] = df["PID"]

# Verificar
print("Número de muestras:", df["PID"].nunique())
print(df[["PID_original", "PID"]].drop_duplicates().head(20))

# Guardar archivo corregido
df.to_csv(
    "datos_50_FA_50_NOFA_PID_consecutivos.csv",
    index=False
)

print("Archivo guardado: datos_50_FA_50_NOFA_PID_consecutivos.csv")
```
